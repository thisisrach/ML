Aim: Understand and analyse the concept of Regression algorithm techniques.
 Program: Implement the non-parametric Locally Weighted Regression algorithm in order to fit data points. Select appropriate data set for your experiment and draw graphs.
 
 LWR: Each point in the dataset is assigned a weight based on its distance from the point of interest. Closer points have higher weights, and distant points have lower weights. Common weighting functions include Gaussian kernels.

```python

import numpy as np
import matplotlib.pyplot as plt

def locally_weighted_regression(x_query, X_train, y_train, tau=0.1):
    m = X_train.shape[0]
    weights = np.exp(-np.sum((X_train - x_query) ** 2, axis=1) / (2 * tau * tau))
    W = np.diag(weights)
    theta = np.linalg.inv(X_train.T @ W @ X_train) @ (X_train.T @ W @ y_train)
    prediction = x_query @ theta
    return prediction

# Generate synthetic data for demonstration
np.random.seed(0)
X_train = np.linspace(0, 10, 50)
y_train = np.sin(X_train) + np.random.normal(0, 0.1, X_train.shape[0])

# Query points
X_query = np.linspace(0, 10, 100)

# Set bandwidth parameter (tau)
tau = 0.5

# Perform Locally Weighted Regression
predictions = []
for xq in X_query:
    x_query = np.array([1, xq])  # Adding bias term
    prediction = locally_weighted_regression(x_query, np.c_[np.ones(X_train.shape[0]), X_train], y_train, tau)
    predictions.append(prediction)

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, color='blue', label='Training data')
plt.plot(X_query, predictions, color='red', label='Locally Weighted Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Locally Weighted Regression')
plt.legend()
plt.grid(True)
plt.show()
```

```python
Small tau (0.1): The regression will be very sensitive to local variations, leading to a fit that follows the noise closely.

Moderate tau (0.5): A balanced fit that captures the overall trend while still responding to local changes.

Larger tau (1.0): The regression will be smoother, focusing more on the general trend and less on local noise.

Very large tau (2.0): The regression will be very smooth, potentially underfitting the data if the local variations are significant.
```
